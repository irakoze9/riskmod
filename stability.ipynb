```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def ode_model(t, y, beta, alpha, theta, sigma, gamma1, gamma2):
    U, E, D, R= y

    dU = -beta * U * D + alpha * E + theta * R
    dE= beta * U * D - (alpha + sigma) * E
    dD= sigma * E - (gamma1 + gamma2) * D
    dR= gamma1 * D - theta * R
    return [dU, dE, dD, dR]

# Parameters
beta = 0.05
alpha = 0.05
theta = 0.05
sigma = 0.05
gamma1 = 0.05
gamma2 = 0.05

# Initial conditions
U0, E0, D0, R0= 40, 30, 20, 10
y0 = [U0, E0, D0, R0]
R0 = (U0 * sigma * beta) / ((alpha + sigma) * (gamma1 + gamma2))
# Time span for simulation
t_span = (0, 80)

# Perform numerical integration
sol = solve_ivp(ode_model, t_span, y0, args=(beta, alpha, theta, sigma, gamma1, gamma2), t_eval=np.linspace(0, 80, 1000))
print (R0)
# Plot the results
plt.plot(sol.t, sol.y[0], label='U')
plt.plot(sol.t, sol.y[1], label='E')
plt.plot(sol.t, sol.y[2], label='D', color='red')
plt.plot(sol.t, sol.y[3], label='R')
plt.xlabel('Time')
plt.ylabel('Initial number of banks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the SIR model
def sir_model(S, I, R, beta, gamma):
    dS = -beta * S * I
    dI = beta * S * I - gamma * I
    dR = gamma * I
    return dS, dI, dR

# Parameters
gamma = 0.1
beta_values = np.linspace(0.01, 0.5, 500)
I_values = []

# Initial conditions
S0 = 0.99
I0 = 0.01
R0 = 0.0

# Create an array of R0 values
for beta in beta_values:
    S = S0
    I = I0
    R = R0
    I_temp = []

    # Time step and total time
    dt = 0.01
    total_time = 100

    for t in range(int(total_time / dt)):
        dS, dI, dR = sir_model(S, I, R, beta, gamma)
        S += dS * dt
        I += dI * dt
        R += dR * dt
        I_temp.append(I)

    I_values.append(max(I_temp))  # Record the maximum I value

# Plot the bifurcation diagram
plt.figure(figsize=(10, 6))
plt.plot(beta_values, I_values, 'b.', markersize=0.5)
plt.xlabel('R0 (Basic Reproduction Number)')
plt.ylabel('Max Infected Population (I)')
plt.title('Bifurcation Diagram for SIR Model')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# SEIR model equations
def seir_model(S, E, I, R, beta, sigma, gamma, N):
    dS = -beta * S * I / N
    dE = (beta * S * I / N) - (sigma * E)
    dI = sigma * E - (gamma * I)
    dR = gamma * I
    return dS, dE, dI, dR

# Parameters
N = 1000  # Total population
sigma = 1/5  # Inverse of incubation period
gamma = 1/7  # Inverse of infectious period
beta_values = np.linspace(0.01, 1.0, 500)  # Range of beta values
I_values = []

# Initial conditions
S0 = N - 1
E0 = 1
I0 = 0
R0 = 0

# Create an array of beta values
for beta in beta_values:
    S = S0
    E = E0
    I = I0
    R = R0
    I_temp = []

    # Time step and total time
    dt = 0.01
    total_time = 100

    for t in range(int(total_time / dt)):
        dS, dE, dI, dR = seir_model(S, E, I, R, beta, sigma, gamma, N)
        S += dS * dt
        E += dE * dt
        I += dI * dt
        R += dR * dt
        I_temp.append(I)

    I_values.append(max(I_temp))  # Record the maximum I value

# Plot the bifurcation diagram
plt.figure(figsize=(10, 6))
plt.plot(beta_values/gamma, I_values, 'b', markersize=0.5)
plt.xlabel('Beta (Transmission Rate)')
plt.ylabel('Max Infected Population (I)')
plt.title('Bifurcation Diagram for SEIR Model')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# SEIR model equations
def model(U, E, D, R, beta, alpha, sigma, gamma1, gamma2, theta):
    dU = -beta * U * D + alpha * E + theta * R
    dE = beta * U * D - (alpha + sigma) * E
    dD = sigma * E - (gamma1 + gamma2) * D
    dR = gamma1 * D - theta * R
    return [dU, dE, dD, dR]

# Parameters
N = 100  # Total population
alpha = 0.02
sigma = 0.02
gamma1 = 0.2
gamma2 = 0.1
theta = 0.1
beta_values = np.linspace(0.0, 0.04, 500)  # Range of beta values
D_values = []

# Initial conditions
U0 = N - 1
E0 = 1
D0 = 0
R0 = 0

# Define R_values as an array
R_values = (beta_values * sigma * N) / ((alpha + sigma) * (gamma1 + gamma2))

# Create arrays to store U, E, D, and R
U_values = []
E_values = []
D_values = []

# Time step and total time
dt = 0.01
total_time = 100

for beta in beta_values:
    U = U0
    E = E0
    D = D0
    R = R0
    D_temp = []

    for t in range(int(total_time / dt)):
        dU, dE, dD, dR = model(U, E, D, R, beta, alpha, sigma, gamma1, gamma2, theta)
        U += dU * dt
        E += dE * dt
        D += dD * dt
        R += dR * dt
        D_temp.append(D)

    D_values.append(max(D_temp))  # Record the maximum D value

# Plot the bifurcation diagram
plt.figure(figsize=(10, 6))
plt.xlim(0, max(R_values))
plt.ylim(0, max(D_values))
for i, D in enumerate(D_values):
    if R_values[i] < 1:
        plt.plot(R_values[i], D, 'g.', markersize=0.5)
    else:
        plt.plot(R_values[i], D, 'r.', markersize=0.5)


plt.xlabel('Basic Reproduction Number (R0)')
plt.ylabel('Max D Value')
plt.title('Bifurcation Diagram for SEIR Model')
plt.grid(True)
plt.show()
```
