```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def ode_model(t, y, beta, alpha, theta, sigma, gamma1, gamma2):
    U, E, D, R= y

    dU = -beta * U * D + alpha * E + theta * R
    dE= beta * U * D - (alpha + sigma) * E
    dD= sigma * E - (gamma1 + gamma2) * D
    dR= gamma1 * D - theta * R
    return [dU, dE, dD, dR]

# Parameters
beta = 0.05
alpha = 0.05
theta = 0.05
sigma = 0.05
gamma1 = 0.05
gamma2 = 0.05

# Initial conditions
U0, E0, D0, R0= 40, 30, 20, 10
y0 = [U0, E0, D0, R0]
R0 = (U0 * sigma * beta) / ((alpha + sigma) * (gamma1 + gamma2))
# Time span for simulation
t_span = (0, 80)

# Perform numerical integration
sol = solve_ivp(ode_model, t_span, y0, args=(beta, alpha, theta, sigma, gamma1, gamma2), t_eval=np.linspace(0, 80, 1000))
print (R0)
# Plot the results
plt.plot(sol.t, sol.y[0], label='U')
plt.plot(sol.t, sol.y[1], label='E')
plt.plot(sol.t, sol.y[2], label='D', color='red')
plt.plot(sol.t, sol.y[3], label='R')
plt.xlabel('Time')
plt.ylabel('Initial number of banks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def ode_model(t, y, beta, alpha, theta, sigma, gamma1, gamma2):
    U, E, D, R = y

    dUdt = -beta * U * D + alpha * E + theta * R
    dEdt = beta * U * D - (alpha + sigma) * E
    dDdt = sigma * E - (gamma1 + gamma2) * D
    dRdt = gamma1 * D - theta * R
    

    return [dUdt, dEdt, dDdt, dRdt]

# Parameters
beta = 0.05
alpha = 0.05
theta = 0.05
sigma = 0.05
gamma1 = 0.05
gamma2 = 0.05


# Varying values of beta
beta_values = np.linspace(0.0, 0.08, 100)  # Adjust the range and number of points as desired

# Initial conditions
U0, E0, D0, R0 = 40, 30, 20, 10
y0 = [U0, E0, D0, R0]

# Calculate R0 and D values for each beta
R0_values = []
D_values = []

for beta in beta_values:
    # Calculate R0
    R0 = (U0 * sigma * beta) / ((alpha + sigma) * (gamma1 + gamma2))
    R0_values.append(R0)

    # Perform numerical integration to get D
    t_span = (0, 100)
    sol = solve_ivp(ode_model, t_span, y0, args=(beta, alpha, theta, sigma, gamma1, gamma2), t_eval=np.linspace(0, 100, 1000))
    D_values.append(sol.y[2][-1])  # Extract the final value of D

# Split the data into two regions based on R0 values
R0_less_than_one = [R0 for R0, D in zip(R0_values, D_values) if R0 < 1]
D_less_than_one = [D for R0, D in zip(R0_values, D_values) if R0 < 1]

R0_greater_than_or_equal_one = [R0 for R0, D in zip(R0_values, D_values) if R0 >= 1]
D_greater_than_or_equal_one = [D for R0, D in zip(R0_values, D_values) if R0 >= 1]
D_values = np.linspace(0.0, 30, 1)
# Plot the results
plt.plot(R0_less_than_one, D_less_than_one,color='red', label='unstable')
plt.plot(R0_greater_than_or_equal_one, D_greater_than_or_equal_one, color ='green', label='stable')
plt.axvline(x=1, color='k', linestyle='--')

plt.xlabel('Basic Reproduction Number ($S_0$)')
plt.ylabel('Distressed banks')
plt.title('Effect of Distress')
plt.legend()
plt.grid(True)
plt.show()
```
